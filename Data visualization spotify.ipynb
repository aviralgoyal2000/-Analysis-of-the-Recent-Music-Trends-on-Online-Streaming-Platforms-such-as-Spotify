```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set()
%matplotlib inline
```

```python
spotify = pd.read_csv('data.csv', encoding='cp1252')
spotify.drop(spotify.columns[0], axis=1, inplace=True)
```

```python
spotify.describe()
```

```python
spotify.info()
```

```python
by_year=spotify.groupby('year').mean()
by_year
```

```python
sns.lineplot(data=by_year["pop"])
```

```python
sns.lineplot(data=by_year["acous"])
```

```python
sns.lineplot(data=by_year["spch"])
```

```python
plt.figure(figsize=(16,20))
sns.countplot(y='top genre', data=spotify)
```

```python
sns.heatmap(spotify.corr(), cmap='magma')
```

```python
sns.countplot(x='year', data=spotify)
```

```python
df=spotify['artist'].value_counts()
df=(df.head(15)).sort_values(ascending=True)
plt.figure(figsize=(8,8))
df.plot(kind='barh',position=1)
```
